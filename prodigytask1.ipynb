```python
import numpy as nu
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('population.csv')
```

```python
data=data.head(20)
```

```python
x=nu.array(data['State'])
y=nu.array(data['Males'])
x_index=nu.arange(len(x))
plt.style.use( 'seaborn-v0_8-white')
width=0.25
plt.figure(figsize=(6, 6))
plt.bar(x_index-width,y,width=0.25,label='male')
plt.bar(x_index,y,width=0.25,label='female')
plt.legend()
plt.xticks(ticks=x_index,labels=x,rotation=90)
plt.grid(True)
plt.tight_layout()
plt.savefig('barchart.png', dpi=300, bbox_inches='tight')
plt.show()
```
